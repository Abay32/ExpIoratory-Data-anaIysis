# Seaborn 


```python
import seaborn as sns 
df = sns.load_dataset('tips')
df.head()
df.dtypes
```

 # Evaluate the correation between the features 

```python
df.corr()
```

# Heatmap plot of the correlation 

```python
sns.heatmap(df.corr()) 
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind='reg')
```

```python
# pair plot (scatter plot) one feature is correated with another feature 
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex') #cassifing with sex
```

```python
sns.distplot(df['tip'], kde='false')
```

```python
# Categorica explotary data analysis 
sns.countplot('sex', data=df) # sex category
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(df['sex'])
```

```python
# Box plot 
sns.boxplot(x='total_bill',y='day', hue='smoker', data=df) # categorizing the smokers 
```

```python
# the other visualization is the violin plot 
sns.violinplot(df['total_bill'], df['day'])
```
